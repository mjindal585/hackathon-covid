```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('Test_dataset.csv')
```

```python
df
```

```python
df.columns
```

```python
df = df[['people_ID',  'Designation', 'Name','Region', 'Gender', 'Married',
       'Children', 'Occupation', 'Mode_transport', 'cases/1M', 'Deaths/1M',
       'comorbidity', 'Age', 'Coma score', 'Pulmonary score',
       'cardiological pressure', 'Diuresis', 'Platelets', 'HBB', 'd-dimer',
       'Heart rate', 'HDL cholesterol', 'Charlson Index', 'Blood Glucose',
       'Insurance', 'salary', 'FT/month']]
```

```python
df
```

```python
temp = {i : j for j , i in enumerate(set(df['Region'].unique()))}
for i,j in enumerate(df['Region']):
    df['Region'][i] = temp[j]
for i,j in enumerate(df['Gender']):
    if j == 'Male':
        df['Gender'][i]=0
    else:
        df['Gender'][i]=1
for i,j in enumerate(df['Married']):
    if j == 'YES':
        df['Married'][i]=1
    else:
        df['Married'][i]=0
temp = {i : j for j , i in enumerate(set(df['Occupation'].unique()))}
for i,j in enumerate(df['Occupation']):
    df['Occupation'][i] = temp[j]
temp = {i : j for j , i in enumerate(set(df['Mode_transport'].unique()))}
for i,j in enumerate(df['Mode_transport']):
    df['Mode_transport'][i] = temp[j]
temp = {i : j for j , i in enumerate(set(df['comorbidity'].unique()))}
for i,j in enumerate(df['comorbidity']):
    df['comorbidity'][i] = temp[j]
temp = {i : j for j , i in enumerate(set(df['Pulmonary score'].unique()))}
for i,j in enumerate(df['Pulmonary score']):
    df['Pulmonary score'][i] = temp[j]
temp = {i : j for j , i in enumerate(set(df['cardiological pressure'].unique()))}
for i,j in enumerate(df['cardiological pressure']):
    df['cardiological pressure'][i] = temp[j]

```
